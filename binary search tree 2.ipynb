```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)


def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)


def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")


def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val == key:
            return root
        elif root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root


def search(root, num, parent):
    if root is None:
        print("The given value {} is not found in the tree!".format(num))
        return 0

    if root.val == num:
        if parent is None:
            print("The given value {} is found".format(num))
            print("There is no parent node to the given node.")
            return 0
        else:
            print("The given value {} is found".format(num))
            print("The parent node is: {}".format(parent.val))
            return 0
    elif num < root.val:
        parent = root
        search(root.left, num, parent)
    elif num >= root.val:
        parent = root
        search(root.right, num, parent)
    else:
        print("The given value {} is not found in the tree!".format(num))


def minimum_node(root):
    if root.left is not None:
        return minimum_node(root.left)
    else:
        return root


def delete_node(node, value):
    if node is None:
        return None
    else:
        if value < node.val:
            node.left = delete_node(node.left, value)
        elif value > node.val:
            node.right = delete_node(node.right, value)
        else:
            if node.left is None and node.right is None:
                node = None
            elif node.left is None:
                node = node.right
            elif node.right is None:
                node = node.left
            else:
                temp = minimum_node(node.right)
                node.val = temp.val
                node.right = delete_node(node.right, temp.val)
    return node


def leaf_nodes(root):
    if root is None:
        pass
    else:
        if root.left is None and root.right is None:
            print(root.val)
        else:
            leaf_nodes(root.left)
            leaf_nodes(root.right)


def main():
    root = None
    done = False
    while not done:
        print()
        print("Select The Operation You Would Like to Perform:")
        print("1. Print The Nodes of the Tree.")
        print("2. Insert Nodes in the Tree.")
        print("3. Search for a value in the Tree.")
        print("4. Delete a Node from the Tree.")
        print("5. Print The Lead Nodes of the Tree")
        print("6. Exit")
        print()
        choice = int(input("Enter your Choice: "))
        if choice == 2:
            key = input("Enter the Value (A,B,C,D): ")
            key = key.split(",")
            for i in key:
                root = insert(root, int(i))
            continue
        elif choice == 1:
            print()
            print("Select The Traversal Order")
            print("1. Inorder")
            print("2. Preorder")
            print("3. Postorder")
            print()
            val = int(input("Enter Your Choice: "))
            if val == 1:
                print()
                print("The Inorder Traversal is: ")
                inorder(root)
                print()
            elif val == 2:
                print()
                print("The Preorder Traversal is: ")
                preorder(root)
                print()
            elif val == 3:
                print()
                print("The Postorder Traversal is: ")
                postorder(root)
                print()
            continue
        elif choice == 3:
            print()
            val = int(input("Enter the value to be searched: "))
            search(root, val, None)
            continue
        elif choice == 4:
            val = int(input("Enter the Value you want to Delete: "))
            delete_node(root, val)
            print()
            print("Deletion completed! {} was Deleted!".format(val))
        elif choice == 5:
            print()
            print("The Leaf Nodes are: ")
            leaf_nodes(root)
            continue
        elif choice == 6:
            print()
            print("Thank You!")
            done = True
        else:
            print()
            print("Invalid Choice!")
            continue


if __name__ == "__main__":
    main()
```

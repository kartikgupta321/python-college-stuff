```python
#operator overloading and dunder and special method
class Employee:
    no_of_leaves=9
    def __init__(self,aname,asalary,arole):
        self.name=aname
        self.salary=asalary
        self.role=arole
    def details(self):
        return f"name is {self.name}.salary is {self.salary} and role is {self.role}"
    def change_leaves(cls,newleaves):
        cls.no_of_leaves=newleaves
    def __add__(self,other):
        return self.salary+other.salary
    def __truediv__(self,other):
        return self.salary/other.salary
    def __repr__(self):
        return f"Employee('{self.name}',{self.salary}, '{self.role}')"
    def __str__(self):
        return  f"name is {self.name}.salary is {self.salary} and role is {self.role}"
emp1=Employee('who',456,'dont')
emp2=Employee('wo',46,'yo')
# print(emp1)
```

```python
#abstract base class
from abc import ABC, abstractmethod
class shape(ABC):
    @abstractmethod
    def printarea(self):
        return 0
class rectangle(shape):
    type='rectangle'
    sides=4
    def __init__(self):
        self.lenght=6
        self.breadth=7
    def printarea(self):
        return self.lenght*self.breadth
rect1=rectangle()
# print(rect1.printarea())
```

```python
# exercise 8 soldier prettify folder
# C:\Users\csc\Desktop\exercise
import os
# path=input('enter string- ')
# file=input('file for file exclusion- ')
# format=input('enter format- ')
def soldier(path,file,format):
    os.chdir(f'{path}')
    with open(f'{file}','r') as f:
        a=f.read()
        s=a.split()
    for item in os.listdir(path):
        if item in s:
            continue
        elif(item.find(format)!=-1):
            continue
        else:
            os.rename(item,item.capitalize())
#             os.rename(item,item.lower())
soldier(path,file,format)
```

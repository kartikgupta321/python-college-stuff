```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class Linkedlist:
    def __init__(self):
        self.head=None
    def search(self,k):
        temp=self.head
        if (temp!=None):
            while(temp!=None):
                if(temp.data==k):
                    print('\nvalue found')
                    break
                temp=temp.next
            if(temp==None):
                print('\nval not found')
        while 1:
            if(temp!=None):
                choice=int(input('\nenter 1 for insertion,2 for deletion after search,else exit- '))
                if(choice==1):
                        val=int(input('enter value for insertion- '))
                        new=Node(val)
                        ptr=self.head
                        while(ptr!=None) and (ptr.data!=k):
                            ptr=ptr.next
                        new.next=ptr.next
                        ptr.next=new
                        self.print()
                elif(choice==2):
                    ptr=self.head
                    preptr=self.head
                    while(ptr!=None) and (ptr.data!=k):
                        preptr=ptr
                        ptr=ptr.next
                    if(ptr.next!=None):
                        preptr.next.next=ptr.next.next
                    else: print('Not possible ')
                    self.print()
                else:
                    break
            else:
                break
    def print(self):
        ptr=self.head
        while(ptr!=None):
            print(f'-->{ptr.data}',end='')
            ptr=ptr.next
    def insertatend(self,k):
        new=Node(k)
        if(self.head==None):
            self.head=new
            return
        else:
            last=self.head
            while(last.next):
                last=last.next
            last.next=new
l1=Linkedlist()
l1.insertatend(10)
l1.insertatend(20)
l1.insertatend(30)
l1.print()
l1.search(30)
l1.print()
```
